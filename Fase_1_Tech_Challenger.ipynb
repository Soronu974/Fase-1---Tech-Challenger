```python
#Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carrega a base de dados
df = pd.read_csv("desafio_nps_fase_1.csv")

# Função responsável por classificar os clientes de acordo
def classificar_nps(nota):
    if nota >= 9:
        return "Promotores"
    elif nota >= 7:
        return "Neutros"
    else:
        return "Detratores"

# Cria uma nova coluna contendo a categoria do cliente
df["categoria_nps"] = df["nps_score"].apply(classificar_nps)

# Conta quantos clientes existem em cada categoria
contagem = df["categoria_nps"].value_counts()

plt.figure(figsize=(6,4))

# Gera um gráfico de barras com a distribuição do NPS
contagem.plot(kind="bar", color=["green", "gold", "red"])

plt.title("Distribuição do NPS")
plt.ylabel("Quantidade de Clientes")

# Mantém os rótulos do eixo X na horizontal
plt.xticks(rotation=0)

plt.tight_layout()

plt.show()
```

```python
# Exibe a quantidade de pedidos realizados em cada região

plt.figure(figsize=(7,4))

df["customer_region"].value_counts().plot(kind="bar")

plt.title("Pedidos por Região")
plt.ylabel("Quantidade")

plt.tight_layout()

plt.show()
```

```python
# Calcula a correlação entre todas as variáveis numéricas
# e a variável alvo (nps_score)

corr = (
    df.corr(numeric_only=True)["nps_score"]
    .drop("nps_score")
    .sort_values()
)

plt.figure(figsize=(8,6))

# Exibe as correlações em um gráfico horizontal
corr.plot(kind="barh")

plt.title("Correlação das variáveis com o NPS")

plt.tight_layout()

plt.show()
```

```python
# Analisa como o número de reclamações influencia
# a distribuição do NPS dos clientes

sns.boxplot(
    data=df,
    x="complaints_count",
    y="nps_score"
)

plt.title("Número de reclamações x NPS")

plt.show()
```

```python
# Seleciona as cinco variáveis com maior correlação positiva
# em relação ao NPS

top = corr.sort_values().tail(5)

plt.figure(figsize=(8,5))

# Exibe os fatores mais associados ao aumento da satisfação
top.plot(kind="barh", color="green")

plt.title("Principais fatores associados ao aumento do NPS")

plt.tight_layout()

plt.show()
```
